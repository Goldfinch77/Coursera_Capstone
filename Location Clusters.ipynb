```python
latlng_df = pd.read_csv('Geospatial_Coordinates.csv')
```

```python
# define a function to get latitude and longitude from latlng_df using postal codes
def getLatLong(postal_code):
    """The function returns a numpy array of size (1,2) containing latitude and longitude 
    for a given postal code"""
    arr = latlng_df[latlng_df['Postal Code']==postal_code][['Latitude','Longitude']].values
    return arr    
```

```python
#use the function getLatLong to obtain Latitude and Longitude information and add to the original dataframe
Toronto_zipcodes['Latitude'] = Toronto_zipcodes[['Postal Code']].applymap(lambda x : getLatLong(x)[0][0])
Toronto_zipcodes['Longitude'] = Toronto_zipcodes[['Postal Code']].applymap(lambda x : getLatLong(x)[0][1])
```

Filter the database to those Boroughs which contain 'Toronto' in their name

```python
Toronto_zipcodes = Toronto_zipcodes[Toronto_zipcodes['Borough'].str.contains('Toronto')].reset_index(drop=True)
```

```python
# define a function to get nearby venues to a neighborhood
def getNearbyVenues(row):
    
    lat = Toronto_zipcodes.loc[row,'Latitude']
    long = Toronto_zipcodes.loc[row,'Longitude']
    
    url = 'https://api.foursquare.com/v2/venues/explore?\
           &client_id={}\
           &client_secret={}\
           &v={}\
           &ll={},{}\
           &radius={}\
           &limit={}'.format(
           CLIENT_ID,
           CLIENT_SECRET,
           VERSION,
           lat,
           long,
           RADIUS,
           LIMIT)
    results = requests.get(url).json()
    n_nby_venues = json_normalize(results['response']['groups'][0]['items'])
    
    string = 'venue.categories'
    
    if string in n_nby_venues.columns:
        n_nby_venues['type'] = n_nby_venues['venue.categories'].apply(lambda x : x[0]['name'])
        n_nby_venues['venue_lat'] = n_nby_venues['venue.location.lat']
        n_nby_venues['venue_long'] = n_nby_venues['venue.location.lng']
        n_nby_venues = n_nby_venues[['venue.name','type','venue_lat','venue_long']]
        n_nby_venues['PostCode'] = Toronto_zipcodes.loc[row,'Postal Code']
        n_nby_venues['Neighbourhood'] = Toronto_zipcodes.loc[row,'Neighbourhood']
        n_nby_venues['Nborhood_latitude'] = Toronto_zipcodes.loc[row,'Latitude']
        n_nby_venues['Nborhood_longitude'] = Toronto_zipcodes.loc[row,'Longitude']
    
        return n_nby_venues
    else:
        return None
    
```

```python
nby_venues = pd.DataFrame()
#loop through for each nborhood and find venues for them
for row in range(0,len(Toronto_zipcodes)):
    n_nby_venue = getNearbyVenues(row)
    nby_venues = pd.concat([nby_venues,n_nby_venue],ignore_index=True)
nby_venues.shape
```

# KMeans clustering

 We will group the nbrhoods into similar clusters based on the venues in the neighborhood

```python
from sklearn.cluster import KMeans
```

```python
# define our input features
X = nby_venues_grpd.drop(['Nborhood_latitude','Nborhood_longitude'],axis=1)
```

```python
X.head()
```

```python
k = 3 # no. of labels
kmean = KMeans(n_clusters=k,random_state=1)
```

```python
kmean.fit(X)
```

```python
labels = kmean.labels_
```

```python
labels
```

```python
#insert the label column in the nby_venues_grpd database
nby_venues_grpd.insert(0,'labels',value=labels)
```

```python
nby_venues_grpd.head()
```

# Plotting

 Make a map of Toronto using Folium and put the markers for nborhoods based on their labels

```python
import folium
from geopy import Nominatim
```

```python
# find the latitude and longitude for Toronto using Geopy
address = 'Toronto,Canada'
geolocator = Nominatim(user_agent='Toronto_explorer')
location = geolocator.geocode(address)
T_latitude = location.latitude
T_longitude = location.longitude
```

```python
T_latitude
```

```python
T_longitude
```

```python
# draw map of Toronto using above coordinates
Toronto_map = folium.Map(location=[T_latitude,T_longitude],zoom_start=12)
```

```python
Toronto_map
```

```python
# place the markers for neighborhoods in the map based on the labels
colors = ['Red','Blue','Green']
#loop through for each label
for i in range(0,3):
    venues_by_label = nby_venues_grpd[nby_venues_grpd['labels']==i].iloc[:,0:3]
    #loop through each nborhood in the particular label
    for lat,long,nbrhood in zip(
        venues_by_label['Nborhood_latitude'],
        venues_by_label['Nborhood_longitude'],
        venues_by_label.index):
        
            label = "{}".format(nbrhood)

            folium.CircleMarker(
            [lat,long],
            radius=3,
            popup=label,
            color=colors[i],
            parse_html=True).add_to(Toronto_map)

```
